```r
read.table('datosgrafico.at') -> d
```

```r
E <- d$V1
S0 <- d$V2
S <- d$V3
density <- d$V4
```

```r
#png('densidad.png')
library(astro)
aplot(density, E, pch = 19, type='o',ylim=c(0,0.8),xlab=c('Densidad Local'),ylab=c('Fraccion de galaxias'))
points(density, S0, pch =8, type='o')
points(density, S, pch=1,type='o')
legend(1.4,.8, pch=c(19,8,1), c('Elipticas','Lenticulares','Espirales + Irregulares'))
#dev.off()
```
