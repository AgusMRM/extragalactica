```python
import numpy as np
import matplotlib.pylab as plt
```

```python
t=np.loadtxt('hist.dat')
```

```python
def schechter(M, alpha, fi0, M0):
    #alpha = -0.759277761
    #fi0   =4.32358503  
    #M0    = -19.8867416
    fi    = fi0 - 0.4*(M-M0)*(alpha+1) - 10**(-0.4*(M-M0))/np.log(10)
    return fi
```

```python
y=[]
x=[]
for i in np.arange(-22.5,-15.5,.1):
    y.append(schechter(i,-0.759277761,4.32358503 ,  -19.8867416 ))
    x.append(i)
```

```python
plt.figure(figsize=(8,8))
plt.rc('text',usetex=True)
plt.plot(t[:,1],np.log10(t[:,0]),linestyle='none',marker='^',color='red')
plt.xlabel(r'$M{r}$',fontsize=20)
plt.ylabel(r'$log_{10}\Phi(M_{r})[Mpc^{-3}mag^{-1}] $',fontsize=20)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.plot(x,y,color='darkblue')
plt.legend(('datos','ajuste'),loc=4,prop={'size': 16})

plt.savefig('/home/agustin/facultad/extragalactica/tp3/LFgalaxies')
```
