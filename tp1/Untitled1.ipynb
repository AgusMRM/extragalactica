```python
import numpy as np
import matplotlib.pylab as plt
import scipy.stats as sp
from ipywidgets import interact
```

```python
t=np.loadtxt('resultados.dat')
```

```python
c=t[:,13]
Mr=t[:,6]
ur=t[:,14]
r50=t[:,16]
```

```python
#filtro por tardias ( c < 2.55) y tempranas (c > 2.55)
mr_tem=[]
r50_tem=[]
mr_tar=[]
r50_tar=[]
for i in range(len(t)):
    if (c[i] > 2.55):
        mr_tem.append(Mr[i])
        r50_tem.append(r50[i])
    if (c[i] <= 2.55):
        mr_tar.append(Mr[i])
        r50_tar.append(r50[i])
```

```python
plt.plot(r50_tem,mr_tem,linestyle='none',marker=',',color='r')
plt.plot(r50_tar,mr_tar,linestyle='none',marker=',',color='b')
plt.semilogx()
plt.ylim(-22.5,-14)
```

```python
mr_tem=[]
r50_tem=[]
mr_tar=[]
r50_tar=[]
for i in range(len(t)):
    if (ur[i] < 2.22):
        mr_tem.append(Mr[i])
        r50_tem.append(r50[i])
    if (ur[i] >= 2.22):
        mr_tar.append(Mr[i])
        r50_tar.append(r50[i])
```

```python
plt.plot(r50_tem,mr_tem,linestyle='none',marker=',',color='r')
plt.plot(r50_tar,mr_tar,linestyle='none',marker=',',color='b')
plt.semilogx()
plt.ylim(-22.5,-14)
```

```python
#realizo un ajuste lineal
np.polyfit(np.log10(r50_tem),mr_tem,1,full='true')
```

```python
p_tem = np.polyfit(np.log10(r50_tem),mr_tem,1)
r50_tem=np.array(r50_tem)
mr_tem=np.array(mr_tem)
ajuste = p_tem[0]*np.log10(r50_tem)+p_tem[1]
plt.plot(r50_tem,mr_tem,linestyle='none',marker='.',color='b')
plt.plot(r50_tem,ajuste,color='yellow',linestyle='-')
plt.semilogx()
plt.ylim(-22.5,-15)
```
